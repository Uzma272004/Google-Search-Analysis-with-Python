```python
import pandas as pd
from pytrends.request import TrendReq
import matplotlib.pyplot as plt
```

```python
!pip install pytrends
```

```python
pytrends = TrendReq(hl='en-US', tz=360)
```

```python
kw_list = ["Cloud Computing"]
```

```python
pytrends.build_payload(
    kw_list,
    cat=0,
    timeframe='today 12-m'
)
```

```python
data = pytrends.interest_over_time()
data.head()
```

```python
data["Cloud Computing"].plot(
    title="Search Interest Over Time"
)
plt.xlabel("Date")
plt.ylabel("Interest")
plt.show()
```

```python
region_data = pytrends.interest_by_region()
region_data = region_data.sort_values(
    by="Cloud Computing",
    ascending=False
)
region_data.head()
```

```python
region_data.reset_index().plot(
    x="geoName",
    y="Cloud Computing",
    kind="bar",
    figsize=(12,6)
)
plt.title("Regional Interest Analysis")
plt.show()
```
